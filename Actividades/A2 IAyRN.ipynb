```python

#Algoritmo que invierta el sentido de una lista de N cantidades
def invertir_lista():
   
    n = int(input("Ingresa el número de elementos de la lista: "))

   
    lista = []
    for i in range(n):
        elemento = input(f"Ingresa el elemento {i+1}: ")
  
        lista.append(elemento)
    print("La lista:", lista)
    
    lista_invertida = lista[::-1]
    print("La lista invertida:", lista_invertida)


invertir_lista()





```
